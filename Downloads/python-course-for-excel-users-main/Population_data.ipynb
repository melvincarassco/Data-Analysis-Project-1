```python
#lets import pandas 
import pandas as pd
```

```python
#Creted a dataframe called raw_data from the csv file.
raw_data = pd.read_csv('population_total.csv')
```

```python
# We need to drop the null values in the following data set
# Two ways to save either overide or add inplace=True
raw_data.dropna(inplace=True)
```

```python
#Created a pivoted table
pivoted_table = raw_data.pivot(index="year",columns="country",values="population")
```

```python
#Selected only few countries for our analysis
pivoted_table = pivoted_table[['United States','India','Indonesia']]
```

```python
#Plotted a line plot
pivoted_table.plot(kind='line')
```

```python
#Enhanced the line plot to make it more descriptive
from turtle import title


pivoted_table.plot(kind='line',xlabel='Year',ylabel='Population',title="Population from 1955 to 2020",figsize=(8,4))
```

```python
#Only using a particular year to plot a bar graph
population_2020 = pivoted_table[pivoted_table.index.isin([2020])]
population_2020
```

```python
 #Performed a transpose to change rows to columns
 population_2020 = population_2020.T
 population_2020
```

```python
#Plotted a bar graph with labels
from turtle import color


population_2020.plot(kind='bar',color='orange',xlabel='Year',ylabel='Population',title='Popilation - 2020')
```

```python
#Renamed the column name from integer to a string
population_2020.rename(columns={2020:'2020'},inplace=True)
population_2020
```

```python
#Plotted a pie chart with the above data
population_2020.plot(kind='pie',y='2020')
```
